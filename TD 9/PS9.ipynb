## Question b)

```python
K = 0.035
LIBOR = np.array([4.228 , 2.791 , 3.067 , 3.728 , 3.728 , 4.051 , 4.051 , 4.199 , 4.199 , 4.45 , 4.45 , 4.626 , 4.626 , 4.816 , 4.816 , 4.96 , 4.96 , 5.088 , 5.088])/100
delta = 1/2
T_n = np.arange(2,11)
```

```python
N = 100
M = 19
lettre = 100
Delto = 1/N
alpha = np.array([[0.00] * N]*M)
H = np.array([[0.00] * N]*M)
Pi = np.array([[0.00] * lettre]*len(T_n))

for t in range(len(T_n)):
    for k in range(lettre):
        B_star = np.array([1.0] * M)
        for m in range(M):
            H[m][0] = np.log(LIBOR)[m]
            for i in range(N-1):
                sig_m = V_m[m] * np.exp(-beta * (T_i[m] - T_n[t]*i*Delto))      
                for j in range(m):
                    alpha[m][i] = alpha[m][i] + delta * np.exp(H[j][i])/(1+delta * np.exp(H[j][i]))*sig_m - 1/2 * sig_m **2
                H[m][i+1] = H[m][i] + alpha[m][i] * Delto + sig_m * np.random.normal(1) * np.sqrt(Delto)
            B_star[m] = (1+ delta * np.exp(H[m-1][m-1]))*B_star[m-1]
            if T_i[m] >= T_n[t]:
                Pi[t][k] = Pi[t][k] + delta*np.maximum(np.exp(H[m][N-1]) - K,0)
        Pi[t][k] = Pi[t][k]/B_star[2*(t+2)]
            #import pdb; pdb.set_trace()
```
